```
# Import widgets
# from ipywidgets import widgets, interactive, interact
# import ipywidgets as widgets
# from IPython.display import display
from IPython.html import  widgets
from IPython.html.widgets import interact
from plotly.offline import iplot

date_series_columns = transaction_date['transactionDate'].astype(str)
# print(date_series_columns, date_series_columns[0])
date_series_columns = np.concatenate([np.array(['all']), date_series_columns.to_numpy()])
# print(date_series_columns, date_series_columns[0])
t_dates = widgets.Dropdown(
    options=date_series_columns,
    value=date_series_columns[0],
    description='日期:'
)


ids = community['name']

series_ids = widgets.Dropdown(
    options=ids,
    value=ids[0],
    description='小区名:'
)
# plotly.offline.init_notebook_mode()
def plot_data(c_id, t_date):
  print(t_date)

  df = ret.loc[ret['name'] == c_id]
  if t_date != 'all':
    df = df.loc[df['transactionDate'] == t_date]

  transaction_date = df.groupby(['transactionDate'], as_index=True)['transactionDate'].count() \
                             .reset_index(name='count') \
                             .sort_values(['count'], ascending=False)
  unit_mean = df.groupby(['transactionDate'], as_index=True)['unitPrice'].mean()

  count = transaction_date.count()['count']

  split = 25 
  plt.figure(figsize=(30,6))
  total = math.ceil(count/split)
  for i in tqdm(range(total)):
    plt.subplot(total, 1, i + 1)
    
    tmp = transaction_date[i * split:(i+1) * split]
    trans_date = tmp['transactionDate'].apply(lambda x: x.strftime('%Y-%m-%d'))
    mean_price = [unit_mean[d] for d in trans_date]
    # print(tmp)
    # print(unit_mean)
    # print(mean_price, unit_mean[tmp['transactionDate']], trans_date)

    fig =sns.barplot(x=trans_date, y=tmp['count'].to_list())

    fig2 = fig.twinx()
    fig2.plot(trans_date, mean_price)
    # fig2 = sns.lineplot(x=trans_date, y=mean_price)
    for item in fig.get_xticklabels():
      item.set_rotation(45)
    # plt.title(c_id + '_' + str(i))

  plt.tight_layout()
  plt.show()



interact(plot_data, c_id=series_ids, t_date=t_dates)

# w = interactive(
#     plot_data,
#     c_id=series_ids
# )
# display(w)
```
